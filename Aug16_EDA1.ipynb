```python
# Importing the country code dataset
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# Getting the final dataframe
final_df = df.merge(df_country, on='Country Code', how='left')
final_df.head()
```

```python
#  how='left' : keeping all columns on left
```

```python
#Getting column names
print(final_df.columns)
```

```python
# Getting proportion of values from country column
final_df['Country'].value_counts()
```

```python
# Data Visualization ->
```

```python
'''
Question ->
Try to plot the pie chart for top 3 countries having
the highest number of restaurants in the given dataset.
'''
# Solution
country_names = final_df['Country'].value_counts().sort_values(ascending = False)[:3].index
country_values = final_df['Country'].value_counts().sort_values(ascending = False)[:3].values

#Plot
plt.figure(figsize=(7,7))
plt.pie(country_values,labels=country_names,autopct='%2f%%',shadow='True')
plt.title('Top 3 country having highest number of restaurants')
plt.legend()
plt.show()
```

```python
'''
Conclusions ->
(1) Majority of the restaurents are from India
(2) USA and UK are followed by that having around 4.64% and 0.87% restaurents.
'''
```

```python
#Ratings
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings.head()
```

```python
#Bar plot
plt.figure(figsize=(15,5))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
plt.title('Count of Aggregate Ratings')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(15,5))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',palette=['blue','red','orange','yellow','green','black'],data=ratings)
plt.title('Count of Aggregate Ratings')
plt.show()
```

```python
"""
   Conclusions ->
   Apart from '0.0 Aggregate Rating' the bar graph is following a bell shaped plot
"""
```

```python
# Count Plot
plt.figure(figsize=(15,5))
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','black'])
plt.title('Count of Aggregate Ratings')
plt.show()
```

```python
'''
   Conclusions ->
   Orange is the highest followed by Red and Yellow, Green, Dark Green and last is White
'''
```

```python
# Finding the proportion of countries that have given 0 ratings
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename(columns={0:'Frequency'})
```

```python
# Question: Draw bar chart and pie chart for the above data
# Solution
country_names = ['Brazil','India','United Kingdom','United States']
frequency_values = [5,2139,1,3]

#Plot
plt.figure(figsize=(7,7))
plt.pie(frequency_values,labels=country_names,autopct='%2f%%',shadow='True')
plt.title('Proportion of countries that have given 0 ratings')
plt.legend()
plt.show()

#Plot
plt.figure(figsize=(12,12))
sns.barplot(x=country_names,y=frequency_values)
plt.title('Proportion of countries that have given 0 ratings')
plt.show()
```

```python
'''
Conclusions: India is the major contributor to 0 ratings, and the other countriesâ€™ contributions are negligible.
'''
```

```python
# Aggregate Rating Wise Country Wise Frequency
final_df.groupby(['Aggregate rating', 'Country']).size().reset_index().rename(columns = {0 : 'Frequency'}).head()
```

```python
'''
Question: -> Try to draw clustered bar plot from the above data
'''
# Solution:
df= final_df.groupby(['Aggregate rating', 'Country']).size().reset_index().rename(columns = {0 : 'Frequency'}).head()
sns.barplot(x = 'Country', y = 'Aggregate rating',data=df,hue='Frequency')
plt.title('Aggregate Rating Wise Country Wise Frequency')
plt.show()

```

```python
'''
   The dataset is heavily skewed towards India, especially for 0.0 aggregate ratings.
'''
```

```python
# Countries focusing on online deliveries
final_df[final_df['Has Online delivery'] == 'Yes'].groupby('Country').size().reset_index().rename(columns = {0 : 'Frequency'})
```

```python
'''
Question: Draw a pie chart for the above data
'''
#Solution->
country_names = ['India','UAE']
frequency_values = [2423,28]

#Plot
plt.figure(figsize=(7,7))
plt.pie(frequency_values,labels=country_names,autopct='%2f%%',shadow='True')
plt.title('Countries focusing on online deliveries')
plt.legend()
plt.show()
```

```python
'''
Conclusion: Online deliveries are primarily concentrated in India, with UAE playing only a very small role.
'''
```

```python
#Pie Chart for top 5 cities
city_values = final_df['City'].value_counts().sort_values(ascending=False)[:5].values
city_index = final_df['City'].value_counts().sort_values(ascending=False)[:5].index

#Plot
plt.figure(figsize=(9,9))
plt.pie(city_values,labels=city_index,autopct='%2f%%',shadow='True')
plt.title('Top 5 cities having highest number of restaurants')
plt.legend()
plt.show()
```

```python
'''
Conclusion: New Delhi dominates the restaurant scene, followed by Gurgaon and Noida, while Faridabad and Ghaziabad contribute very little.
'''
```
